```python
import pandas as pd
import numpy as np

df = pd.read_csv('../data/raw/Dataset.csv', header=2)
```

```python
df.head(10)
```

```python

print("Original Columns:", df.columns.tolist())

new_columns = [
    'Commodity_Group',  # Col 0
    'Commodity',        # Col 1
    'Variety',          # Col 2
    'MSP',              # Col 3
    'Price_Today',      # Col 4
    'Price_1DayAgo',    # Col 5
    'Price_2DaysAgo',   # Col 6
    'Arrival_Today',    # Col 7
    'Arrival_1DayAgo',  # Col 8
    'Arrival_2DaysAgo'  # Col 9
]

# Assign these new names to the first 10 columns
df.columns = new_columns + list(df.columns[10:])
df = df[new_columns]

print("\nNew Standardized Columns:", df.columns.tolist())
df.head(3)
```

```python
# sometimes the prices and arrivals will be left blank either becuase the crops didnt arrive or data wasnt taken so
# we will keeep it as NaN values becuase we cannot guess it just like that
numeric_cols = [
    'MSP', 'Price_Today', 'Price_1DayAgo', 'Price_2DaysAgo',
    'Arrival_Today', 'Arrival_1DayAgo', 'Arrival_2DaysAgo'
]

for col in numeric_cols:
    df[col] = df[col].replace('-', np.nan)
    df[col] = df[col].astype(str).str.replace(',', '')
    df[col] = pd.to_numeric(df[col], errors = 'coerce')

print(df.isnull().sum())

print(df.head(15))
```

```python
df.to_csv("../data/processed/day1_cleaned.csv", index=False)
print("Completed saving")
```
